```python
#handling missing values(using mean ans scitkit-learn)
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values = np.nan,strategy='mean')
ds[['Age','Salary']] = si.fit_transform(ds[['Age','Salary']])
```

```python
#encoding categorical column(onehotencoding)
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False)
encoded_data = ohe.fit_transform(ds[['Country']])
encoded_data
```

```python
# creating columns
column_names = ohe.get_feature_names_out(['Country'])
encoded_df = pd.DataFrame(encoded_data,columns = column_names)
encoded_df
```

```python
#joining original dataset and new dataframe
ds = pd.concat([ds,encoded_df],axis =1)
ds
```

```python
ds.info()
```

```python
# drop actual categorical country column
ds = ds.drop(columns = ['Country'])
ds
```

```python
#lable encoding 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ds['Purchased'] = le.fit_transform(ds['Purchased'])
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
ds[['Age','Salary']] = sc.fit_transform(ds[['Age','Salary']])
```

```python
#dividing dataset into independent (input)and dependent(output)variables

x = ds[['Age','Salary','Country_France','Country_Germany','Country_Spain']]
y = ds[['Purchased']]
```
